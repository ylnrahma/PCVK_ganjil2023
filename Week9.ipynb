7. Buatlah program implementasi operasi opening dengan Structuring Element
berbentuk square 3 x 3 tanpa dan dengan menggunakan library morphology dari
openCV, sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan gambar
fingerprint.png)





*   Operasi Tanpa Library



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca image untuk operasi erosi
img2 = cv2.imread('/content/drive/MyDrive/PCVK/images/fingerprint.png', 0)

# Acquire size of the image
m, n = img2.shape

# Define the structuring element
k = 3
SE = np.ones((k, k), dtype=np.uint8)
constant = (k - 1) // 2

# Define new image
imgErode = np.zeros((m, n), dtype=np.uint8)

# Erosion without using inbuilt cv2 function for morphology
for i in range(constant, m - constant):
    for j in range(constant, n - constant):
        temp = img2[i - constant:i + constant + 1, j - constant:j + constant + 1]
        product = temp * SE
        imgErode[i, j] = np.min(product)

# Read the image for dilation
img2_2 = imgErode

# Acquire size of the image
p, q = img2_2.shape

# Define new image to store the pixels of the dilated image
imgDilate = np.zeros((p, q), dtype=np.uint8)

# Define the structuring element
SED = np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])
constant1 = 1

# Dilation operation without using the inbuilt cv2 function
for i in range(constant1, p - constant1):
    for j in range(constant1, q - constant1):
        temp = img2_2[i - constant1:i + constant1 + 1, j - constant1:j + constant1 + 1]
        product = temp * SED
        imgDilate[i, j] = np.max(product)

plt.subplot(131), plt.imshow(img2, cmap="gray"), plt.title('Citra Awal')
plt.subplot(132), plt.imshow(imgErode, cmap="gray"), plt.title('Citra Hasil Erosi')
plt.subplot(133), plt.imshow(imgDilate, cmap="gray"), plt.title('Citra Hasil Opening')
plt.show()
```



*   Operasi Dengan Library



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Acquire size of the image
m, n = img2.shape

# Menampilkan citra asli
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.imshow(img2, cmap="gray")
plt.title('Citra Awal')

# Define the structuring element
kernel = np.ones((3, 3), np.uint8)

# Implementasi operasi erosi dengan library morphology
erosion = cv2.erode(img2, kernel, iterations=1)

plt.subplot(132)
plt.imshow(erosion, cmap="gray")
plt.title('Citra Hasil Erosi')

# Implementasi operasi opening dengan library morphology
opening = cv2.morphologyEx(img2, cv2.MORPH_OPEN, kernel)

plt.subplot(133)
plt.imshow(opening, cmap="gray")
plt.title('Citra Hasil Opening')

plt.show()
```

8. Buatlah program implementasi operasi closing dengan Structuring Element berbentuk square 3 x 3 tanpa dan dengan menggunakan library morphology dari openCV, sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan gambar
fingerprint.png)





*   Operasi Tanpa Library



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca image
img3 = cv2.imread('/content/drive/MyDrive/PCVK/images/fingerprint.png', 0)

# Acquire size of the image
m, n = img3.shape

# Define the structuring element
k = 3
SE = np.ones((k, k), dtype=np.uint8)
constant = (k - 1) // 2

# Define new image for dilation
imgDilate = np.zeros((m, n), dtype=np.uint8)

# Define new image for closing
imgClose = np.zeros((m, n), dtype=np.uint8)

# Dilation operation without using inbuilt cv2 function
for i in range(constant, m - constant):
    for j in range(constant, n - constant):
        temp = img3[i - constant:i + constant + 1, j - constant:j + constant + 1]
        product = temp * SE
        imgDilate[i, j] = np.max(product)

# Erosion operation without using inbuilt cv2 function for morphology
for i in range(constant, m - constant):
    for j in range(constant, n - constant):
        temp = imgDilate[i - constant:i + constant + 1, j - constant:j + constant + 1]
        product = temp * SE
        imgClose[i, j] = np.min(product)

plt.figure(figsize=(12, 6))

plt.subplot(131), plt.imshow(img3, cmap="gray"), plt.title('Citra Awal')
plt.subplot(132), plt.imshow(imgDilate, cmap="gray"), plt.title('Citra Hasil Dilasi')
plt.subplot(133), plt.imshow(imgClose, cmap="gray"), plt.title('Citra Hasil Closing')

plt.show()
```

9. Buatlah program implementasi operasi dilasi dengan Structuring Element berbentuk Cross 3 x 3 dan 5 x 5, sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan gambar j.png)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca citra
img4 = cv2.imread('/content/drive/MyDrive/PCVK/images/j.png', 0)

# Acquire size of the image
m, n = img4.shape

# Menampilkan citra asli
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(img4, cmap="gray")
plt.title('Citra Awal')

# Implementasi operasi dilasi dengan Structuring Element berbentuk Cross 3x3
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
dilation = cv2.dilate(img4, kernel, iterations=1)

print("SED = Cross 3")

plt.subplot(122)
plt.imshow(dilation, cmap="gray")
plt.title('Citra Hasil Dilasi Cross 3x3')
plt.show()

# Tambahkan pengaturan subplot yang benar untuk hasil Cross 5x5
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(img4, cmap="gray")
plt.title('Citra Awal')

# Implementasi operasi dilasi dengan Structuring Element berbentuk Cross 5x5
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))
dilation = cv2.dilate(img4, kernel, iterations=1)

print("SED = Cross 5")

plt.subplot(122)
plt.imshow(dilation, cmap="gray")
plt.title('Citra Hasil Dilasi Cross 5x5')

plt.show()
```
